```python
import numpy as np
import pandas as pd
from sklearn.cluster import SpectralClustering
USERNAME = 'pleasedshibe1'
similarity_matrix = np.load(f'../{USERNAME}_similarity_matrix.npy')
df = pd.read_csv(f'../{USERNAME}finished.csv')
#round?
print(similarity_matrix)
df.drop('Unnamed: 0', axis=1, inplace=True)


distmtx = []
for i, row in enumerate(similarity_matrix):
    distrow = []
    for j, value in enumerate(row):
        if i == j:
            distrow.append(0)
        else:
            distrow.append(1 - value)
    distmtx.append(distrow)
    
```

```python
len(similarity_matrix)

from sklearn.cluster import DBSCAN
dbscan_clusters = DBSCAN(eps=0.5, min_samples=5, metric='precomputed', metric_params=None, algorithm='auto', leaf_size=30, p=None, n_jobs=None).fit_predict(distmtx)
```

```python
df['Dcluster'] = dbscan_clusters
```
